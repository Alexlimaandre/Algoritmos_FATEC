1) Calcule a soma da seguinte série, onde N é um valor digitado pelo usuário (acredite que é um número múltiplo de 5):

	1/5 + 1/10 + 1/15 + 1/20 + 1/25 + ... + 1/N

Use repetição.
Exemplo (N digitado foi 15):
1/5 + 1/10 + 1/15 = 0,2 + 0,1 + 0,0667 = 0,3667




2) Dada uma lista de números inteiros, crie três listas:

* menores que 10
* entre 10 e 20 (inclusive)
* maiores que 20

**Exemplo:**

**Entrada:** [5, 10, 15, 25, 8]

**Saída:**

Menores que 10: [5, 8]

Entre 10 e 20: [10, 15]

Maiores que 20: [25]


3) Verifique se uma lista está em ordem estritamente crescente.

**Exemplo:**

**Entrada:** [1, 2, 3, 4] → True

**Entrada:** [1, 2, 2, 3] → False (a lista não é estritamente crescente se existirem elementos iguais: 2 e 2)

**Entrada:** [5, 4, 3] → False




4) Um programa recebe o valor de várias vendas na forma de uma lista. Ao final, deve informar:

* Total vendido
* Quantas vendas foram acima de R$ 100

**Exemplo:**

**Entrada:** [50, 200, 150, 80]

**Saída:**

**Total:** 480

**Vendas acima de 100:** 2
