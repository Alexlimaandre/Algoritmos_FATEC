1) Calcule a soma da seguinte série, onde N é um valor digitado pelo usuário (acredite que é um número múltiplo de 5):

	1/5 + 1/10 + 1/15 + 1/20 + 1/25 + ... + 1/N

Use repetição.
Exemplo (N digitado foi 15):
1/5 + 1/10 + 1/15 = 0,2 + 0,1 + 0,0667 = 0,3667



```python
n = int(input("Digite um número: "))
soma = 0

for i in range(0, n+1, 5):
  if(i == 0):
    continue
  else:
    soma += 1/i

print(soma)
```


2) Dada uma lista de números inteiros, crie três listas:

* menores que 10
* entre 10 e 20 (inclusive)
* maiores que 20

**Exemplo:**

**Entrada:** [5, 10, 15, 25, 8]

**Saída:**

Menores que 10: [5, 8]

Entre 10 e 20: [10, 15]

Maiores que 20: [25]

```python
lista = [5,10,15,25,8]
lista_m_d = []
lista_d_e_v = []
lista_m_vinte = []

for i in lista:
    if i < 10:
      lista_m_d.append(i)
    elif i <= 20:
      lista_d_e_v.append(i)
    else:
      lista_m_vinte.append(i)
print(f"Menor{lista_m_d}\nEntre{lista_d_e_v}\nMaior{lista_m_vinte}")
```


3) Verifique se uma lista está em ordem estritamente crescente.

**Exemplo:**

**Entrada:** [1, 2, 3, 4] → True

**Entrada:** [1, 2, 2, 3] → False (a lista não é estritamente crescente se existirem elementos iguais: 2 e 2)

**Entrada:** [5, 4, 3] → False



```python
lista00 = [0,1,2,9,2,3,4]
lista_ordenada = sorted(lista00)
lista_e_ord = []
res = False

for i in lista_ordenada:
  if i not in lista_e_ord:
    lista_e_ord.append(i)

if lista_e_ord == lista00:
  res = True
else:
  res = False

print(res)
```


4) Um programa recebe o valor de várias vendas na forma de uma lista. Ao final, deve informar:

* Total vendido
* Quantas vendas foram acima de R$ 100

**Exemplo:**

**Entrada:** [50, 200, 150, 80]

**Saída:**

**Total:** 480

**Vendas acima de 100:** 2

```python
lista_vendas = [50,200,150,80]
total = sum(lista_vendas)
qtd_m_100 = 0

for i in lista_vendas:
  if i > 100:
    qtd_m_100 += 1

print(f"{total}\n{qtd_m_100}")
```
