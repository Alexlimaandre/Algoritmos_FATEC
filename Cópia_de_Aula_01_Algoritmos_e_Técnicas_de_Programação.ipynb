**Um último lembrete:**  Quando fazemos mais de uma conta  (operação aritmética) dentro de uma expressão, em que ordem elas são realizadas?

Lembre de **P E M D A S** (é nessa ordem!!):

-**P** de parêntesis,

-**E** de exponenciação e potenciação,

-**M** de multiplicação,

-**D** de divisão,

-**A** de adição,

-**S** de subtração

**DICA bem legal!!** Se você quiser encontrar automaticamente a cotação do dolar ou de outras moedas, você poderia consultar algum site que forneça esta informação em tempo real. Veja o exemplo de código abaixo:

```
import requests, json
from datetime import datetime

hoje = datetime.now()
hoje_formatado = hoje.strftime("%m-%d-%Y")  # Formato "MM-DD-YYYY"

requisicao = requests.get('https://brasilapi.com.br/api/cambio/v1/cotacao/USD/' + hoje_formatado)
if requisicao.status_code != 200:
    print('Erro de conexão')
    exit()
cotacao = json.loads(requisicao.text)
preco_dolar = cotacao['cotacoes'][0]['cotacao_compra']
print(preco_dolar)
```

10) Faça um programa que calcule a área de um triângulo com base em seus três lados (fórmula de Herão ou Hierão). Sendo a, b e c os lados:
$$
p = \frac{a+b+c}{2},
area = \sqrt{ p*(p-a)*(p-b)*(p-c) }
$$

```
# Entrada de dados
a = float(input("Entre com a medida do lado A: "))
b = float(input("Entre com a medida do lado B: "))
c = float(input("Entre com a medida do lado C: "))

# Processamento dos dados (contas)
# Parêntesis Exponenciação_ou_potenciação Multiplicação Divisão Adição Subtração
p = (a+b+c)/2
area = (p*(p-a)*(p-b)*(p-c)) ** 0.5

# Saída dos dados (resultados)
print('A área do triângulo mede', area)
```

11) **(DESAFIO)** Crie um código que descubra aproximadamente quantos dias você já viveu, com base no dia, mês e ano de nascimento, e dia, mês e ano de hoje.


Agora, só por curiosidade, vamos mostrar o jeito que usa a biblioteca de datas do python:

```
from datetime import date

dh = int(input("Dia hoje: "))
mh = int(input("Mês hoje: "))
ah = int(input("Ano hoje: "))
dn = int(input("Dia nascimento: "))
mn = int(input("Mês nascimento: "))
an = int(input("Ano nascimento: "))
t1 = date(year = ah, month = mh, day = dh)
t2 = date(year = an, month = mn, day = dn)

t3 = t1 - t2

print("Dias vividos =", t3.days)
```

14) Para calcular a área de um pentágono regular, podemos dividi-lo internamente em triângulos, como mostra a imagem. Se determinarmos as medidas da base e da altura de um destes triângulos, é possível calcular a área total da figura.

Pensando nisso, faça um programa que recebe a medida de um dos lados do pentágono e calcula sua área total.

```
import math
# Entrada de dados
lado = float(input("Entre com a medida do lado do pentágono: "))

# Processamento
area = (5*lado**2) / (4*math.tan(math.radians(36)))

# Saída de dados
print(f"O pentágono tem área: {area:.2f}")
```
