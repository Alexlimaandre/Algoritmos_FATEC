5) Ache o maior e o menor valor dentre os elementos de uma matriz (faça do jeito *raiz*, ou seja, *for* dentro de *for*.. *if*... Do jeito *Nutella* foi feito abaixo).

````python
# Do jeito Nutella é assim:
import numpy as np
A = np.array( [[3, 6, 4, 2],
               [1, 2, 3, 4],
               [2, 4, 5, 8]] )

print(A.max())
print(A.min())
````

````python
# Jeito raiz (para praticar)
# usando matrizes nativas do python:
import numpy as np
A = np.matrix( [[3, 6, 4, 2],
                [1, 2, 3, 4],
                [2, 4, 5, 8]] )

menor = A.item((0,0))
maior = A.item((0,0))
for l in range(A.shape[0]):  # A.shape[0] significa o número de linhas de A
  for c in range(A.shape[1]):  # A.shape[1] significa o número de colunas de A
    if A.item((l,c)) > maior: maior = A[l,c]
    elif A.item((l,c)) < menor: menor = A[l,c]
print(maior)
print(menor)

# usando matrizes nativas do python:
A = [[3, 6, 4, 2],
    [1, 2, 3, 4],
    [2, 4, 5, 8]]

menor = A[0][0]
maior = A[0][0]
for l in range(len(A)):  # len(A) significa o número de linhas de A
  for c in range(len(A[0])):  # len(A[0]) significa o número de colunas de A
    if A[l][c] > maior:
      maior = A[l][c]
    elif A[l][c] < menor:
      menor = A[l][c]
print(maior)
print(menor)
````

Resolvam o algoritmo do jeito *raiz*, ou seja, *for* dentro de *for*.. soma cumulativa... *if*... Do jeito *Nutella* foi feito abaixo:

````python
import numpy as np
# Entrada das dimensões da matriz
dimensoes = input().split()
L = int(dimensoes[0])
C = int(dimensoes[1])

# Entrada dos valores da matriz
linhas = ''
for l in range(L):
  linhas = linhas + ' ' + input()

valoresInteiros = []
for x in linhas.split():
  valoresInteiros = valoresInteiros + [ int(x) ]

mat = np.array(valoresInteiros)
mat.resize(L, C)

# Solução do jeito Nutella

somaLinhas = mat.sum(axis=1)
somaColunas = mat.sum(axis=0)
maxLinhas = somaLinhas.max()
maxColunas = somaColunas.max()
print(maxLinhas if maxLinhas > maxColunas else maxColunas)
````

#**Funções**

*Funções* são trechos de código que recebem entradas, fazem operações e podem devolver respostas. Escreve-se uma vez e pode-se chamá-la ou executá-la várias vezes.

Podemos identificar trechos que são necessários em muitas partes do sistema e escrever funções para implementá-los. Também há muitas bibliotecas já prontas em Python, com dezenas de funções cada uma.

Um bom exemplo disso é o conjunto de funções matemáticas, como raiz quadrada (sqrt), logaritmo (log, log2, log10), seno (sin), cosseno (cos), tangente (tan).

O uso de funções traz muitas vantagens:
* O código fica mais organizado.
* O código fica mais curto (normalmente).
* O código fica com melhor manutenibilidade (mais fácil de corrigir e manter).
* O código passa a ser mais reutilizável.
```
               --->|------------|---->
      entradas --->|   Função   |----> saídas
               --->|------------|---->
```

1) Crie uma função que calcule o fatorial de um número inteiro.
```
0! = 1
1! = 1
2! = 1*2 = 2
3! = 1*2*3 = 6
4! = 1*2*3*4 = 24
.....
n! = 1*2*3*4*5*....*(n-1)*n
```
Com ajuda dessa função, calcule o número de arranjos simples de n elementos organizados em grupos de p:

$$
A_{n,p} = \frac{n!}{(n-p)!}
$$

````python
def fatorial(x):
  fat=1
  for i in range(1, x+1):
    fat=fat*i

  return fat


n = int(input("Entre com o número de elementos: "))
p = int(input("Entre com o tamanho dos grupos: "))
a = fatorial(n) / fatorial(n-p)
print('Haverá %d arranjos simples' % a)
#print('Haverá', a, 'arranjos simples')
#print(f'Haverá {a:.0f} arranjos simples')
````

2) Usando o algoritmo de [Newton-Raphson](https://pt.wikipedia.org/wiki/M%C3%A9todo_de_Newton%E2%80%93Raphson), implemente uma função que calcule raiz quadrada. Neste [link](https://youtu.be/yneIBwkmbGo) você pode assistir uma aula sobre o algoritmo.

````python
def minhaRaizQuadrada(N):
  candidato = N/2
  diferenca = 1
  while diferenca > 0.000001:
    novo = (candidato*candidato + N)/(2*candidato)
    diferenca = (novo - candidato) if novo > candidato else (candidato - novo)
    candidato = novo
  return novo


x = float(input('Entre com um numero: '))
raiz = minhaRaizQuadrada(x);
print('A raiz quadrada de', x, 'é', raiz)
import math
print('A raiz quadrada (de acordo com a biblioteca matemática) de', x, 'é', math.sqrt(x))
````

Agora vamos criar um algoritmos mais simples para a raiz quadrada, aproximando sucessivamente o valor por um passo de 0.1:

````python
def raizBobinha(n):
  i=1
  while i<n:
    print(f'Estou tentando {i}*{i} que dá {i*i}')
    if i*i > n:
      i = i-1
      break
    i=i+1

  while i<n:
    print(f'Estou tentando {i}*{i} que dá {i*i}')
    if i*i > n:
      i = i-0.1
      break
    i=i+0.1

  while i<n:
    print(f'Estou tentando {i}*{i} que dá {i*i}')
    if i*i > n:
      return i-0.01
    i=i+0.01

x = int(input('Entre com um numero: '))
raiz = raizBobinha(x);
print('A raiz quadrada de', x, 'é', raiz)
import math
print('A raiz quadrada (de acordo com a biblioteca matemática) de', x, 'é', math.sqrt(x))
````

3) Vamos calcular o valor aproximado de PI: $$\pi = 3.14159265358979323846 ....$$

Podemos usar a [*série de Leibniz*](https://pt.wikipedia.org/wiki/F%C3%B3rmula_de_Leibniz_para_%CF%80):

$$
\pi = \frac{4}{1} - \frac{4}{3} + \frac{4}{5} - \frac{4}{7} + \frac{4}{9} - \frac{4}{11} + ... = \sum_{i=0}^{+\infty} \frac{4.(-1)^i}{2.i+1}
$$


````python
def pi(n):
  soma=0
  for i in range(n+1):
    soma = soma + (4*((-1)**i)) / (2*i+1)

  return soma


print('PI =', pi(1000))
print('PI =', pi(10000))
print('PI =', pi(100000))
print('PI =', pi(1000000))
print('PI =', pi(10000000))
````

Uma ideia: Calcular o PI pelo método de [Monte Carlo](https://www.blogcyberini.com/2018/09/calculando-o-valor-de-pi-via-metodo-de-monte-carlo.html):

````python
import random
import math

def sorteia_ponto():
    return (random.uniform(-1, 1),
            random.uniform(-1, 1))

def checa_se_dentro_circulo(ponto):
    x,y = ponto
    distancia = math.sqrt(x**2 + y**2)
    if distancia<=1:
        return True
    else:
        return False

def pi(n):
    i=1
    cont=0
    while i<=n:
        ponto = sorteia_ponto()
        if checa_se_dentro_circulo(ponto):
            cont += 1
        i+=1

    return cont/n*4

print(pi(1))
print(pi(10))
print(pi(100))
print(pi(1000))
print(pi(10000))
print(pi(100000))
print(pi(1000000))
````
