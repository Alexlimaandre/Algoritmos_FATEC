# **Coleções (parte 1)**

Coleções representam formas de armazenar muitos dados na memória do computador durante a execução do programa.

Existem 4 tipos principais de coleções em Python:
* Listas ou vetores (de fato, no mundo Python não se costuma usar o nome *vetor*)
* Tuplas
* Dicionários
* Conjuntos

Cada uma destas tem características que as fazem mais úteis para uma ou outra situação.

#Listas ou vetores
Uma lista é uma variável que pode armazenar vários dados na memória (uma variável simples guarda apenas *um* dado) sequencialmente; para se identificar cada um dos dados usa-se um índice, ou número de ordem: 0 é o primeiro, 1 o segundo, 2 o terceiro, e assim por diante...

Uma analogia que podemos fazer é a seguinte: uma variável simples, como as que já aprendemos, é como uma casa; um vetor poderia ser assemelhado a um edifício com vários apartamentos. Repare que para visitar alguém que more numa casa, precisamos ter o endereço (logradouro e número da casa); só que, para visitarmos outro alguém num apartamento, além do logradouro e número da rua, precisamos saber qual o número do apartamento. Este número pode ser encarado da mesma forma que o índice do vetor.

Como poderíamos calcular a média aritmética das notas da sala?

````
soma = 0
for i in range(len(notas)): # aqui nós usamos índice
  soma = soma + notas[i]

media = soma / len(notas) # len descobre o tamanho da lista
print('Média da sala =', media)
````

De um outro jeito, usando uma abordagem mais simples (o que chamamos de *iteradores*):

````
soma = 0
for n in notas:
  soma = soma + n

media = soma / len(notas) # len descobre o tamanho da lista
print('Média da sala =', media)
````

**Sabe o que é um histograma?**

"O histograma, também conhecido como distribuição de frequências, é a representação gráfica em colunas ou em barras (retângulos) de um conjunto de dados previamente tabulado e dividido em classes uniformes ou não uniformes. A base de cada retângulo representa uma classe. A altura de cada retângulo representa a quantidade ou a frequência absoluta com que o valor da classe ocorre no conjunto de dados para classes uniformes ou a densidade de frequência para classes não uniformes.Importante ferramenta da estatística, o histograma também é uma das chamadas sete ferramentas da qualidade."
[Wikipedia](https://pt.wikipedia.org/wiki/Histograma)

Histograma_wiki1.png

Vamos construir um histograma dos valores presentes na lista/vetor sorteados:

````
histograma = [0]*7  # crio uma lista com 7 zeros: [0, 0, 0, 0, 0, 0, 0]

for i in range(len(sorteados)):
    face = sorteados[i]
    histograma[face] = histograma[face] + 1

for face in range(1,7):  # mostrando o histograma
  porc = int( histograma[face]/len(sorteados) * 100 )
  print(face, porc*'|', porc, '%')
````

````
import math
vet = [5,7,8,4,9,2,0,9.5,4,7.5]

# Média aritmética

# Zerésima abordagem (não será usada NUNCA!!!!)
soma = vet[0] + vet[1] + vet[2] + vet[3] + vet[4] + vet[5] + vet[6] + vet[7] + vet[8] + vet[9]
media = soma / 10
print(f'Média = {media:.2f}')

# Primeira abordagem
soma = 0
for i in range(len(vet)):
  soma = soma + vet[i]
media = soma / len(vet)
print(f'Média = {media:.2f}')

# Segunda abordagem
soma = 0
for valor in vet:
  soma = soma + valor
media = soma / len(vet)
print(f'Média = {media:.2f}')

# Terceira abordagem
media = sum(vet) / len(vet)
print(f'Média = {media:.2f}')

# Desvio padrão

# Zerésima abordagem (não será usada NUNCA!!!!)
soma = (vet[0]-media)**2 + (vet[1]-media)**2 + (vet[2]-media)**2 + (vet[3]-media)**2 + \
  (vet[4]-media)**2 + (vet[5]-media)**2 + (vet[6]-media)**2 + (vet[7]-media)**2 + \
  (vet[8]-media)**2 + (vet[9]-media)**2
desvio = math.sqrt(soma / 10)
print(f'Desvio = {desvio:.6f}')

# Primeira abordagem
soma = 0
for i in range(len(vet)):
  soma = soma + (vet[i]-media)**2
desvio = math.sqrt(soma / len(vet))
print(f'Desvio = {desvio:.6f}')

# Segunda abordagem
soma = 0
for valor in vet:
  soma = soma + (valor-media)**2
desvio = math.sqrt(soma / len(vet))
print(f'Desvio = {desvio:.6f}')
````

5) Os números a seguir vêm de um sensor que mede os batimentos cardíacos. Uma medida é realizada a cada 0,2 segundos.  Escreva um código que leia 20 amostras e diga qual é a frequência cardíaca do paciente. Exemplo:

100, 50, 0, 0, 0, 100, 40, 5, 5, 2, 90, 20, 10, 5, 5, 95, 30, 0, 0, 0

Neste caso acima, a frequência cardíaca é de 60 batimentos. Veja o raciocínio:

Acompanho a sequência até ocorrer um ponto onde o valor não aumenta mais (chamamos de máximo local):
```
100, 50, 0, 0, 0, 100, 40, ...
                   ^-este é um máximo local
```
Descubro quantos pontos separam este de outro ponto que é máximo local:
```
100, 50, 0, 0, 0, 100, 40, 5, 5, 2, 90, 20, ...
                   ^                ^
                   máx. local       máx. local
```
São 5 medidas, portanto 5 x 0,2 segundos = 1 segundo, ou seja, ocorre um batimento por segundo, assim 60 batimentos por minuto!

Para testar seu algoritmo, tente os valores:

Ex. 1) vet = [100,50,0,0,0,100,40,5,5,2,90,20,10,5,5,95,30,0,0,0]

Frequência = 60.0 batimentos por segundo

Ex. 2) vet = [100,50,0,0,100,40,5,2,90,20,5,5,95,30,0,0,90,45,0,0]

Frequência = 75.0 batimentos por segundo

Explicação do resultado acima (exemplo 2):
	4 intervalos * 0.2s = 0.8s de período
$$
	freq = \frac{1}{período}, freq = 1/0.8 = 1.25 bat/s = 1.25 * 60 bat/min = 75 bat/min
$$

````
vet = [100,50,0,0,100,40,5,2,90,20,5,5,95,30,0,0,90,45,0,0]
i=1
while i < len(vet)-1:
  anterior = vet[i-1]
  atual = vet[i]
  posterior = vet[i+1]
  # Descobrir um máximo local
  if anterior < atual > posterior: # o mesmo que if anterior < atual and atual > posterior:
    break
  else: # caso contrário eu continuo procurando
    i=i+1

indiceMaximo1 = i
i=i+1
while i < len(vet)-1:
  if vet[i-1] < vet[i] > vet[i+1]: # Máximo local
    break
  else:
    i=i+1

indiceMaximo2 = i
frequenciaBatMin = 60/(0.2*(indiceMaximo2-indiceMaximo1))
print('Frequência =', frequenciaBatMin, 'batimentos/minuto')
````
